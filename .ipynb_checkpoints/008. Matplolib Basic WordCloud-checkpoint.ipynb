## Pie Chart

- explode는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정  

- autopct는 부채꼴 안에 표시될 숫자의 형식을 지정

````python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=explode)
plt.show()
````

## Stack Plot

- 스택 플롯은 각 부분이 전체를 구성하는 방식을 쉽게 시각화하여 전체 데이터 세트를 보여주는 플롯.

- 스택 플롯의 각 구성 요소는 서로 겹쳐진다.

- 전체 단위뿐만 아니라 단위의 부분 구성을 보여준다.

````python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(5, 3))       
# stacked area chart 생성 
ax.stackplot(yrs, rng + rnd, labels=['Estasia', 'Eurasis', 'Oceania'])   

ax.set_title("Combined debt growth ratio")
ax.legend(loc="upper left")
ax.set_ylabel("Total debt")
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()                                    
````

## multiple subplots

**y = x + noise 인 (x, y) 쌍의 toy data 생성**

````python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)

data = np.column_stack((x, y))
data[:5]
````

### plt.subplots() 는 OOP style 에서 pyplot 을 이용하는 유일한 경우이다. 

````python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.scatter(x, y, marker='o', c='r', edgecolor='b')
ax1.set_title("Scatter: x vs. y")
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.hist(data, bins=10, label=('x', 'y'))
ax2.set_title("Frequency")
ax2.legend()
ax2.yaxis.tick_right()
````

- 동일한 plot 을 함수형 방식으로 그리기

````python
fig = plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(x, y, marker='o', c='r', edgecolor='b')
plt.title("Scatter: x vs. y")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 2, 2)
plt.hist(data, bins=10, label=('x', 'y'))
plt.title("Frequency")
plt.legend()
plt.tick_params(axis='y', which='both', left=False, labelleft=False, labelright=True)
````

## gridspec 을 이용한 subplot customization

- gridspec 패키지를 사용하여 플롯의 여백과 개별 서브 플로트 간의 간격을 쉽게 제어   


- 동일한 그림에 다른 크기의 축을 허용

<img src="empty_gridspec.webp" width="300">
<img src="empty_gridspec_annot.webp" width="300">

````python
from matplotlib import gridspec

plt.figure()
gs = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gs[0, 1:])
side_histogram = plt.subplot(gs[1:, 0])
lower_right = plt.subplot(gs[1:, 1:])

X = np.random.random(1000)
Y = np.random.randn(1000)

lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal')

side_histogram.set_ylabel('Y')
lower_right.set_xlabel("X")

plt.tight_layout()
````

## ticker

### tick formatters
<img src="tickformatter.png" width="500">

- set_major_formatter() : Set the formatter of the major ticker  
- yaxis.get_ticklabels : Get the tick labels as a list of Text instances

````python
import matplotlib.ticker as ticker

samples = 100*np.random.rand(20)

fig, ax = plt.subplots()
ax.plot(samples)

formatter = ticker.FormatStrFormatter('$%1.2f')
ax.yaxis.set_major_formatter(formatter)

for tick in ax.yaxis.get_ticklabels():
    tick.set_color('red')
    tick.set_fontsize(13)

for tick in ax.xaxis.get_ticklabels():
    tick.set_color('blue')
    tick.set_rotation(45)
    tick.set_fontsize(13)
````

## Word Cloud 

`Word Cloud` (텍스트 클라우드 또는 태그 클라우드라고도 함)는 간단한 방식으로 작동합니다. 텍스트 데이터 소스 (예 : 음성, 블로그 게시물 또는 데이터베이스)에 특정 단어가 더 많이 나타날수록 더 크고 굵게 표시됩니다. 

`Word Cloud`는 일반적으로 텍스트 데이터의 고급 분석 및 시각화를 수행하는 데 사용됩니다. *이상한 나라의 엘리스*라는 단편 소설을 분석해 보겠습니다.

````python
# !pip install wordcloud
# !pip install Pillow
````

````python
from wordcloud import WordCloud, STOPWORDS

# open the file and read it into a variable alice_novel
alice_novel = open('data/alice_novel.txt', 'r').read()
````

다음으로 `wordcloud`에서 가져온 불용어를 사용하겠습니다. *set* 함수를 사용하여 중복 불용어를 제거합니다.

````python
stopwords = set(STOPWORDS)
````

단어 구름 개체를 만들고 단어 구름을 생성합니다. 단순화를 위해 소설의 처음 2000 개 단어 만 사용하여 단어 구름을 생성 해 보겠습니다.

````python
# instantiate a word cloud object
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)
````

````python
# display the word cloud
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
````

소설의 처음 2000 개 단어에서 가장 일반적인 단어는 **Alice**, **said**, **little**, **Queen** 등 입니다. **said**은 실제로 유용한 단어가 아닙니다. 이제 불용어에 추가하고 클라우드를 다시 생성 해 보겠습니다.

자주 사용하지 않는 단어를 조금 더 잘 볼 수 있도록 클라우드 크기를 조정 해 보겠습니다.

````python
# add the words said to stopwords
stopwords.add('said') 

# re-generate the word cloud
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
````

`word_cloud` 패키지로 구현할 수있는 또 다른 멋진 기능은 모든 모양의 마스크에 단어를 겹쳐 놓는 것입니다. 

````python
from PIL import Image
alice_mask = np.array(Image.open('alice_mask.png'))
````

````python
fig = plt.figure()
fig.set_figwidth(4) # set width
fig.set_figheight(6) # set height

plt.imshow(alice_mask, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
````

````python
# instantiate a word cloud object
alice_wc = WordCloud(background_color='white', max_words=2000, 
                     mask=alice_mask, stopwords=stopwords)

# generate the word cloud
alice_wc.generate(alice_novel)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc)
plt.axis('off')
plt.show()
````

## 실전 문제

### Data Science 분야에 대한 관심도 조사 결과를 시각화

````python
import pandas as pd

df_survey = pd.read_csv('data/Topic_Survey_Assignment.csv', index_col=0)
df_survey
````

### 1 단계 - 관심도 높은 순서로 정렬

````python
df_survey.sort_values('Very interested')
df_survey
````

### 2 단계 - 응답한 숫자를 전체 응답 숫자에 대한 비율로 전환

- dataframe.divide
    - Get Floating division of dataframe and other, element-wise 

````python
row_sum = df_survey.sum(axis=1)
row_sum
````

````python
df_survey = round(df_survey.divide(row_sum, axis=0) * 100, 2)
df_survey
````

### 3 단계 - 조사 분야별로 bar chart 작성하여 비교

- x 축 : 조사 분야  
- y 축 : 응답 비율  
- column 별로 색 구분하여 bar chart  plot

````python
categories = df_survey.index
categories
````

````python
fig, ax = plt.subplots(figsize=(16, 6))

index = np.arange(len(df_survey))

very_interest = df_survey.iloc[:, 0].values
some_interest = df_survey.iloc[:, 1].values
no_interest = df_survey.iloc[:, 2].values

ax.bar(index, very_interest, width=0.3, color='red', label="Very Interested")

ax.bar(index+0.3, some_interest, width=0.3, color='green', 
       label="somewhat_interested", tick_label=categories)                  

ax.bar(index+0.6, no_interest, width=0.3, color='blue', label="not_interested")

ax.set_title("Percentage of the Respondent's interest in Data Science Area", fontsize=16)
ax.set_xticklabels(categories, rotation=90)

plt.legend(fontsize=11)
````
